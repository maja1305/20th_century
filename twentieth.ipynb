```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
df = pd.DataFrame({
    'id': range(1, 11),
    'value': np.random.randint(10, 100, size=10),
    'country': np.random.choice(['Germany','Serbia','Spain','Italy','France'], size=10)
})
df
```

```python
df['value_plus_5'] = df['value'].apply(lambda x: x + 5)
df
```

```python
df['is_germany'] = df['country'].apply(lambda c: c == 'Germany')

region_map = {
    'Germany': 'EU Central',
    'France': 'EU West',
    'Spain': 'EU South',
    'Italy': 'EU South',
    'Serbia': 'EU Balkans'
}
df['region'] = df['country'].apply(lambda c: region_map.get(c, 'Other'))
df
```
